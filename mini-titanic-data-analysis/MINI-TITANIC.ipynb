What is the percentage of folks that survived?

```python
total_passengers = len(df['SURVIVED'])
total_survived = df['SURVIVED'].sum()
percentage_survived = (total_survived / total_passengers) * 100
print(f"Percentage survived: {percentage_survived:.2f}%")
```
